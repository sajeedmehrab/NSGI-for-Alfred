```python
from pathlib import Path
```

```python
input_data_file = Path("./input_corrected.txt")
output_data_file = Path("./output_corrected.txt")
```

```python
input_data = input_data_file.open('r').readlines()
output_data = output_data_file.open('r').readlines()
```

```python
input_data = [data.strip() for data in input_data]
output_data = [data.strip() for data in output_data]
```

# Task Completion

```python
output_last_steps = []
for data in output_data:
    if data.endswith('<->'):
        last_step = data.split('<->')[-2]
    else:
        last_step = data.split('<->')[-1]
    output_last_steps.append(last_step)
```

```python
# last step matching
input_last_steps = [step.split('<->')[-1].strip() for step in input_data]
```

# Exact Match

```python
count = 0
for inp, out in zip(input_last_steps, output_last_steps):
    inp_cln = inp.replace(",", "").replace(".", "").replace(" ", "").lower()
    out_cln = out.replace(",", "").replace(".", "").replace(" ", "").lower()
    if inp_cln == out_cln:
        count += 1
```

```python
count/len(input_last_steps)
```

```python
import nltk
nltk.download('punkt')
```

```python
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity
from nltk.tokenize import word_tokenize

# Example original and generated sequences
original_sequence = input_last_steps
generated_sequence = output_last_steps

# Tokenize sentences into words
original_tokens = [word_tokenize(step) for step in original_sequence]
generated_tokens = [word_tokenize(step) for step in generated_sequence]

# Load or train a Word2Vec model based on your data
sentences = original_tokens + generated_tokens
model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)

# Keep track of completed last step
completed_last_step = 0
# Calculate average word vectors for the last steps
for orig_step, gen_step in zip(original_tokens, generated_tokens):
    orig_seq = ' '.join(orig_step)
    gen_seq = ' '.join(gen_step)
    print(orig_seq)
    print(gen_seq)
    #print(' '.join(generated_tokens))
    # Calculate average embedding for the words in orig_step and gen_step
    orig_emb = sum([model.wv[word] for word in orig_step]) / len(orig_step)
    gen_emb = sum([model.wv[word] for word in gen_step]) / len(gen_step)
    #print(orig_emb.shape)
    #print(gen_emb.shape)
    
    # Reshape vectors for cosine_similarity function
    orig_emb = orig_emb.reshape(1, -1)
    gen_emb = gen_emb.reshape(1, -1)

    # Calculate cosine similarity
    similarity_score = cosine_similarity(orig_emb, gen_emb)[0][0]

    # Print the similarity score
    #print(f"Cosine Similarity Score: {similarity_score}")
    if similarity_score >= 0.80:
        completed_last_step += 1
    
```
